A PYTHON PROGRAM TO READ A SENTENCE AND DISPLAY HOW MANY
TIMES EACH WORD APPEARS, IGNORING CASE AND PUNCTUATION

```python
import string
text = input("Enter the Text: ").lower()
print("Entered Text:")
print(text)
for char in string.punctuation:
    text = text.replace(char, "")   

words = text.split()
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1
print("\nOutput:")  
print(freq) 
```
