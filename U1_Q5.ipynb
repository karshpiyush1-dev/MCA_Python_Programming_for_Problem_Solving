```python
def probability(favorable_outcomes, total_outcomes):
    return favorable_outcomes / total_outcomes

def probability_of_die_roll():
    total_outcomes = 6
    favorable_outcomes = 1  # Rolling a 3
    prob= probability(favorable_outcomes, total_outcomes)
    print(f"Probability of rolling a 3 on a fair six-sided die: {prob}")

def probability_of_coin_flip():
    total_outcomes = 2
    favorable_outcomes = 1  # Getting heads
    prob = probability(favorable_outcomes, total_outcomes)
    print(f"Probability of getting heads on a fair coin flip: {prob}")

probability_of_die_roll()
probability_of_coin_flip()
```
