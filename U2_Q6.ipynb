Write a program to search for an item in a user-provided list and display
 the position if found, otherwise print “Item not found.”

```python
def search_item_in_list():
    try:
        # Take list input from user
        size = int(input("Enter the number of elements in the list: "))
        if size <= 0:
            print("List size must be a positive integer.")
            return
        
        user_list = []
        for i in range(size):
            element = input(f"Enter element {i+1}: ") 
            user_list.append(element)
        
        # Take the item to search
        search_item = input("Enter the item to search: ") 
        
        # Search for the item
        if search_item in user_list:
            position = user_list.index(search_item) + 1 
            print(f"Item found at position {position}")
        else:
            print("Item not Found")
    
    except ValueError:
        print("Invalid input. Please enter valid data.")

        #Run the function
if __name__ == "__main__": 
        search_item_in_list() 
```
