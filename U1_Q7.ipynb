A PYTHON PROGRAM TO FIND FACTORIAL OF A NUMBER USING RECURSION FUNCTION

```python
def recursion(Num):
    if Num == 1:
        return 1
    else:
        return Num*recursion(Num-1)
Num = int(input("eneter a number:"))
print(f"The Number : {Num}")
print(f"Factorial of {Num} is {recursion(Num)}")
```
