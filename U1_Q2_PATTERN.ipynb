--THE PATTERN PROGRAMS --

RIGHT ANGLE TRIANGLE PATTERN

```python
print("Right Angle Triangle Number Pattern \n")
num=6
for col in range(1, num):
    for row in range(1, col+1):
        print(row, end =" ")
    print() # gives a new line
```

PYRAMID PATTERN

```python
print("Pyramid Pattern")
Num = 4
var =1
for col in range (1, Num +1):
    for row in range (Num-col):
        print(" ", end=" ")
    for row in range(var):
        print("* ", end= "")
    var+=2
    print()
```

DIAMOND PATTERN

```python
n = int(input("Enter an odd number for the diamond size: "))

# upper half
for i in range(1, n+1, 2):
    print((" " * ((n - i) // 2)) + ("*" * i))

# lower half
for i in range(n-2, 0, -2):
    print((" " * ((n - i) // 2)) + ("*" * i))
```

 HOLLOW SQUARE PATTERN


```python
n = int(input("Enter the size of the square: "))

for i in range(n):
    if i == 0 or i == n - 1:
        print("*" * n)              # top or bottom row
    else:
        print("*" + " "*(n-2) + "*") # middle rows
```

STAR SHAPED or X PATTERN

```python
for i in range(3):
    for j in range(2):
        if (i==1) and (j==0 or j==2):
           print("  ", end="")
        else:
           print("*  ",end="")    
           print()    
```
