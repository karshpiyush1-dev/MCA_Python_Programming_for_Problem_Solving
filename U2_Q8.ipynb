A PYTHON PROGRAM THAT READS A TEXT FILE AND COUNTS THE NUMBER OF LINES, WORDS AND CHARACTERS.

```python
with open('sample.txt', 'r') as f:
    lines = f.readlines()

number_of_lines = len(lines)
number_of_words = sum(len(line.split()) for line in lines)
number_of_characters = sum(len(line) for line in lines)

print(f"Lines:", number_of_lines)
print(f"Words:", number_of_words)
print(f"Characters:", number_of_characters)
   
```
