A PYTHON PROGRAM FEATURE A LIST OF EMPLOYEE RECORDS AS DICTIONARIES, SORT THEM BY SALARY AND DISPLAY THE SORTED LIST.

```python
employe_list = [
    {"name":"Phil Salt","Salary":45000},
    {"name":"Virat Kohli","Salary":55000},
    {"name":"Tim David","Salary":40000},
   
    ]
print(f"Original List: {employe_list}\n")

# Sorting the list of dictionaries by Salary in descending order
employe_list.sort(key=lambda x: x['Salary'])
print(f"Sorted List by Salary (Ascending): {employe_list}\n")   
```
